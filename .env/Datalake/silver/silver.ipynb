```python
import pandas as pd
```

```python
# Salvando o arquivo csv em um DataFrame
df = pd.read_csv('../bronze//GasPricesinBrazil_2004-2019.csv', sep=';')

# Exibindo as primeiras linhas do DataFrame
df.head(5)
```

```python
# Excluindo a coluna 'Unnamed: 0'
del df['Unnamed: 0']
```

```python
# Exibindo as tabelas com valores nulos e a quantidade de valores nulos
df.isnull().sum()
```

```python
# alterando toodos os valores nulos para 0
df = df.fillna(0)

df.isnull().sum()
```

```python
df.info()
```

```python
# Convertendo as colunas para o tipo datetime

df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'], errors='coerce')
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'], errors='coerce')
```

```python
# Convertendo as colunas para o tipo float

df['DESVIO PADRÃO REVENDA'] = pd.to_numeric(df['DESVIO PADRÃO REVENDA'], errors='coerce')
df['PREÇO MÁXIMO REVENDA'] = pd.to_numeric(df['PREÇO MÁXIMO REVENDA'], errors='coerce')
df['MARGEM MÉDIA REVENDA'] = pd.to_numeric(df['MARGEM MÉDIA REVENDA'], errors='coerce')
df['COEF DE VARIAÇÃO REVENDA'] = pd.to_numeric(df['COEF DE VARIAÇÃO REVENDA'], errors='coerce')
df['PREÇO MÉDIO DISTRIBUIÇÃO'] = pd.to_numeric(df['PREÇO MÉDIO DISTRIBUIÇÃO'], errors='coerce')
df['DESVIO PADRÃO DISTRIBUIÇÃO'] = pd.to_numeric(df['DESVIO PADRÃO DISTRIBUIÇÃO'], errors='coerce')
df['PREÇO MÍNIMO DISTRIBUIÇÃO'] = pd.to_numeric(df['PREÇO MÍNIMO DISTRIBUIÇÃO'], errors='coerce')
df['PREÇO MÁXIMO DISTRIBUIÇÃO'] = pd.to_numeric(df['PREÇO MÁXIMO DISTRIBUIÇÃO'], errors='coerce')
df['COEF DE VARIAÇÃO DISTRIBUIÇÃO'] = pd.to_numeric(df['COEF DE VARIAÇÃO DISTRIBUIÇÃO'], errors='coerce')
```
