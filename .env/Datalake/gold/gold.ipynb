```python
import pandas as pd
```

```python
data = pd.read_csv('../silver/GasPricesinBrazil_limpo_2004-2019.csv', sep=';')
data.head(5)
```

```python
# Exibindo os valores únicos da coluna 'REGIÃO'
data['REGIÃO'].unique()
```

```python
# FILTRANDO OS DADOS POR REGIÃO
NORTE = data['REGIÃO'] == 'NORTE'
data[NORTE].head(5)
```

```python
# Exibindo os valores únicos da coluna 'ESTADO'
data['ESTADO'].unique()
```

```python
# Salvando os dados filtrado por REGIÃO em um novo arquivo csv
data[data['REGIÃO'] == 'CENTRO OESTE'].to_csv('CENTRO_OESTE.csv', sep=';', decimal=',', index=False)
data[data['REGIÃO'] == 'NORDESTE'].to_csv('NORDESTE.csv', sep=';', decimal=',', index=False)
data[data['REGIÃO'] == 'NORTE'].to_csv('NORTE.csv', sep=';', decimal=',', index=False)
data[data['REGIÃO'] == 'SUDESTE'].to_csv('SUDESTE.csv', sep=';', decimal=',', index=False)
data[data['REGIÃO'] == 'SUL'].to_csv('SUL.csv', sep=';', decimal=',', index=False)
```
